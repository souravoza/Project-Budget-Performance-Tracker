```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
# Replace 'your_file_path' with the actual path to your CSV file.
data_path = '/content/BuildMacro_Dataset.csv'
df = pd.read_csv(data_path)

# Step 2: Check if data loaded correctly
print("First few rows of the dataset:")
print(df.head())  # Print the first few rows to understand the structure

# Step 3: Budget Analysis
def budget_analysis(df):
    # Calculate percentage of budget used
    df['Budget Utilization (%)'] = (df['Cost to Date ($)'] / df['Budget ($)']) * 100
    # Filter projects exceeding 90% budget usage
    exceeding_budget = df[df['Budget Utilization (%)'] > 90]
    print("\nProjects Exceeding 90% of Budget:")
    print(exceeding_budget[['Project Name', 'Budget Utilization (%)']])

# Step 4: Cost Efficiency
def cost_efficiency(df):
    # Calculate cost efficiency for each project
    df['Cost Efficiency ($/Hour)'] = df['Cost to Date ($)'] / df['Hours Spent']
    print("\nCost Efficiency for Each Project:")
    print(df[['Project Name', 'Cost Efficiency ($/Hour)']])

# Step 5: Timeline Management
def timeline_management(df):
    # Find ongoing projects
    ongoing_projects = df[df['Status'] != 'Completed']
    print("\nOngoing Projects Nearing End Date:")
    print(ongoing_projects[['Project Name', 'End Date', 'Status']])

# Step 6: Visualization
def visualize_budget_vs_cost(df):
    # Create a bar chart comparing budget and cost to date
    plt.figure(figsize=(10, 6))
    plt.bar(df['Project Name'], df['Budget ($)'], label='Budget ($)', alpha=0.7)
    plt.bar(df['Project Name'], df['Cost to Date ($)'], label='Cost to Date ($)', alpha=0.7)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Project Name')
    plt.ylabel('Cost ($)')
    plt.title('Budget vs Cost to Date')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Step 7: Run the functions
budget_analysis(df)
cost_efficiency(df)
timeline_management(df)
visualize_budget_vs_cost(df)


```
